# Ejemplos de DataFrames con Pandas

## Ejemplo 1: Información de personas

```python
import pandas as pd

df1 = pd.DataFrame({
    'Nombre': ['Ana', 'Luis', 'Marta', 'Carlos'],
    'Edad': [23, 35, 29, 42],
    'Ciudad': ['Madrid', 'Barcelona', 'Sevilla', 'Valencia']
})
df1
```

## Ejemplo 2: Productos en inventario

```python
df2 = pd.DataFrame({
    'Producto': ['Manzana', 'Plátano', 'Naranja', 'Uva'],
    'Precio': [1.2, 0.8, 1.5, 2.0],
    'Stock': [100, 150, 80, 200]
})
df2
```

## Ejemplo 3: Ventas diarias

```python
df3 = pd.DataFrame({
    'Fecha': pd.date_range('2023-01-01', periods=5, freq='D'),
    'Ventas': [10, 15, 7, 20, 13],
    'Ganancia': [25, 40, 15, 60, 35]
})
df3
```
