# Ejemplos de Concurrencia y Paralelismo en Python

Este notebook muestra la diferencia entre **concurrencia** y **paralelismo** con ejemplos prácticos en Python.

## Concurrencia con `asyncio`
Ejemplo de múltiples tareas concurrentes usando *corutinas*.

```python
import asyncio
import time

async def tarea(nombre, tiempo):
    print(f"⏳ Iniciando {nombre}")
    await asyncio.sleep(tiempo)
    print(f"✅ {nombre} terminada en {tiempo} segundos")

async def main():
    inicio = time.time()
    await asyncio.gather(
        tarea("Tarea 1", 2),
        tarea("Tarea 2", 3),
        tarea("Tarea 3", 1)
    )
    print(f"Tiempo total: {time.time() - inicio:.2f} segundos")

await main()
```

## Paralelismo con `multiprocessing`
Ejemplo de ejecución en paralelo usando múltiples procesos.

```python
from multiprocessing import Process
import os
import time

def trabajo(nombre):
    print(f"🔹 Proceso {nombre} ejecutándose en PID {os.getpid()}")
    time.sleep(2)
    print(f"✅ Proceso {nombre} terminado")

if __name__ == "__main__":
    inicio = time.time()
    
    p1 = Process(target=trabajo, args=("A",))
    p2 = Process(target=trabajo, args=("B",))

    p1.start()
    p2.start()

    p1.join()
    p2.join()

    print(f"Tiempo total: {time.time() - inicio:.2f} segundos")
```

## Diferencias clave
- **Concurrencia**: varias tareas progresan compartiendo tiempo de ejecución (ideal para operaciones de I/O).
- **Paralelismo**: varias tareas se ejecutan realmente al mismo tiempo en distintos núcleos (ideal para cálculos pesados).
