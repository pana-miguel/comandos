# Manejo de paquetes en Python
Este cuaderno explica cómo organizar, crear y usar paquetes en Python.

Incluye ejemplos de:
- creación de módulos
- estructura de paquetes con `__init__.py`
- importaciones absolutas y relativas
- uso de `pip` y entornos virtuales
- instalación y uso de paquetes externos
- buenas prácticas de organización

## 1. ¿Qué es un módulo y qué es un paquete?

- **Módulo**: archivo `.py` que contiene funciones, clases o variables.
- **Paquete**: carpeta que contiene varios módulos y un archivo especial `__init__.py`.

Ejemplo de estructura:
```
mi_paquete/
│── __init__.py
│── modulo1.py
│── modulo2.py
```


## 2. Crear un módulo sencillo

````python
# Creamos un archivo ejemplo llamado 'matematicas.py'
%%writefile matematicas.py

def suma(a, b):
    return a + b

def resta(a, b):
    return a - b
````

## 3. Importar y usar el módulo

````python
import matematicas

print('Suma:', matematicas.suma(5, 3))
print('Resta:', matematicas.resta(5, 3))
````

## 4. Crear un paquete con varios módulos

````python
# Creamos un paquete llamado 'mi_paquete'
!mkdir -p mi_paquete

# Archivo __init__.py (vacío o con inicialización)
%%writefile mi_paquete/__init__.py
# Paquete de ejemplo

# modulo operaciones
%%writefile mi_paquete/operaciones.py
def multiplicar(a, b):
    return a * b

# modulo utilidades
%%writefile mi_paquete/utilidades.py
def saludar(nombre):
    return f'Hola, {nombre}!'
````

## 5. Usar el paquete

````python
from mi_paquete import operaciones, utilidades

print('Multiplicar:', operaciones.multiplicar(4, 5))
print(utilidades.saludar('Miguel'))
````

## 6. Importaciones relativas dentro de un paquete

En `__init__.py` o en otros módulos del paquete, puedes usar:
```python
from .operaciones import multiplicar
from .utilidades import saludar
```
Esto facilita la organización interna.

## 7. Uso de `pip` y entornos virtuales

```bash
# Crear entorno virtual
python -m venv venv

# Activar (Linux/Mac)
source venv/bin/activate

# Activar (Windows)
venv\Scripts\activate

# Instalar un paquete externo
pip install requests
```


## 8. Usar un paquete externo instalado con pip

````python
import requests

resp = requests.get('https://api.github.com')
print('Código de respuesta:', resp.status_code)
print('Headers:', resp.headers.get('content-type'))
````

---

### Notas finales

- Usa `__init__.py` para definir qué se expone al importar un paquete.
- Organiza tus módulos por responsabilidad.
- Usa entornos virtuales para aislar dependencias.
- Publica tus paquetes en **PyPI** usando `setuptools` si quieres compartirlos.
