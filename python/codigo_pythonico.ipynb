# Ejemplos de Código Pitónico
Este cuaderno muestra ejemplos de cómo escribir código más limpio, eficiente y legible en Python.

## 1. List Comprehension

```python
# No pitónico
squares = []
for i in range(10):
    squares.append(i**2)
print(squares)

# Pitónico
squares = [i**2 for i in range(10)]
print(squares)
```

## 2. Uso de enumerate en lugar de range(len(...))

```python
# No pitónico
fruits = ["manzana", "banana", "cereza"]
for i in range(len(fruits)):
    print(i, fruits[i])

# Pitónico
for i, fruit in enumerate(fruits):
    print(i, fruit)
```

## 3. Dict Comprehension

```python
# No pitónico
cuadrados = {}
for i in range(5):
    cuadrados[i] = i**2
print(cuadrados)

# Pitónico
cuadrados = {i: i**2 for i in range(5)}
print(cuadrados)
```

## 4. Uso de with para manejar archivos

```python
# No pitónico
file = open("ejemplo.txt", "w")
file.write("Hola mundo")
file.close()

# Pitónico
with open("ejemplo.txt", "w") as file:
    file.write("Hola mundo")
```

## 5. Uso de any() y all()

```python
# No pitónico
nums = [1, 2, 3, 4, 5]
hay_mayores = False
for n in nums:
    if n > 3:
        hay_mayores = True
        break
print(hay_mayores)

# Pitónico
nums = [1, 2, 3, 4, 5]
print(any(n > 3 for n in nums))

# Comprobar si todos son positivos
print(all(n > 0 for n in nums))
```
