# Anotaciones en Python

En este cuaderno encontrarás una explicación detallada de **todas las anotaciones en Python**, con ejemplos y descripciones.


## Comentarios (#)

Los comentarios en Python se escriben con el símbolo `#`.
Todo lo que esté después de este símbolo en la misma línea será ignorado por el intérprete.

Ejemplo:
```python
# Esto es un comentario
x = 5  # Esto también es un comentario al final de una línea de código
```

## Docstrings (""" """)

Las *docstrings* son cadenas multilínea que se colocan al inicio de módulos, funciones, clases o métodos para documentar su propósito.

Ejemplo:
```python
def sumar(a, b):
    """Esta función recibe dos números
     y retorna su suma"""
    return a + b
```

## Type Hints (anotaciones de tipo)

permiten indicar el tipo de variables, parámetros y valores de retorno de funciones. 
No afectan la ejecución del código, pero ayudan a la documentación y a herramientas como `mypy`.

Ejemplo:
```python
def multiplicar(a: int, b: int) -> int:
    return a * b

edad: int = 25
nombre: str = "Miguel"
```

## Anotaciones en clases dataclass

Las clases con `@dataclass` usan anotaciones de tipo para definir los campos de forma más clara y automática.

Ejemplo:
```python
from dataclasses import dataclass

@dataclass
class Persona:
    nombre: str
    edad: int
```

## Anotaciones en variables locales y globales

Se pueden añadir anotaciones a variables, tanto locales como globales, usando `: tipo`.

Ejemplo:
```python
x: float = 3.14
y: list[int] = [1, 2, 3]
```

## Anotaciones en literales (typing)

El módulo `typing` añade anotaciones más avanzadas como `List`, `Dict`, `Tuple`, `Union`, `Optional`.

Ejemplo:
```python
from typing import List, Dict, Optional

nombres: List[str] = ["Ana", "Juan"]
edades: Dict[str, int] = {"Ana": 25, "Juan": 30}
telefono: Optional[str] = None
```

## Anotaciones con Annotated

`Annotated` permite agregar metadatos a las anotaciones de tipo.

Ejemplo:
```python
from typing import Annotated

def procesar(nombre: Annotated[str, "Debe ser un nombre válido"]) -> None:
    print(nombre)
```

## Anotaciones en variables constantes (Final)

Con `Final`, se indica que una variable no debe reasignarse.

Ejemplo:
```python
from typing import Final

PI: Final = 3.14159
```

## Anotaciones con Literal

`Literal` permite indicar valores exactos que puede tomar una variable o parámetro.

Ejemplo:
```python
from typing import Literal

def mover(direccion: Literal["izquierda", "derecha"]):
    print(f"Moviendo hacia {direccion}")
```

## Anotaciones en funciones genéricas

Con `TypeVar` se pueden crear funciones o clases genéricas que aceptan múltiples tipos.

Ejemplo:
```python
from typing import TypeVar

T = TypeVar("T")

def identidad(x: T) -> T:
    return x
```
